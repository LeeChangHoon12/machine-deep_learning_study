```python
####20164512 이창훈####
####2020 06 03 기계학습 실습####
###Stochastic Gradient Descent(SGD)#

import tensorflow as tf
import numpy as np
```

```python
xy = np.loadtxt('data-03-diabetes.csv',delimiter=',',dtype=np.float32)
```

```python
xy.shape
```

```python
# x_data = xy[:,0:-1]
# y_data = xy[:,[-1]]
# print(x_data.shape)
# print(y_data.shape)

##### data shuffle for tochastic Gradient Descent(SGD)  #####
xp = tf.placeholder(np.float32,shape=[None,8])
yp = tf.placeholder(np.int32,shape=[None,1])

#tf.cast(yp,tf.int32)
y_one_hot = tf.one_hot(yp,depth=2)
y_one_hot = tf.reshape(y_one_hot,[-1,2])

w1 = tf.Variable(tf.random_normal([8,16]))

L1 = tf.nn.relu(tf.matmul(xp,w1))

w2 = tf.Variable(tf.random_normal([16,32]))

L2 = tf.nn.relu(tf.matmul(L1,w2))

w3 = tf.Variable(tf.random_normal([32,2]))
h = tf.matmul(L2,w3)




loss = tf.nn.softmax_cross_entropy_with_logits(logits=h,labels=y_one_hot)
loss = tf.reduce_mean(loss)

train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(loss)

print(loss)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())

sess.run(loss,feed_dict={xp:xy[:,0:8],yp:xy[:,8:]})


traing_epoch = 1000
batch_size = 100

sample_size = xy.shape[0]
 
    
for epoch in range(traing_epoch):
    np.random.shuffle(xy)
    for b in range(0,sample_size,batch_size):
        e = b + batch_size
        if e >= sample_size:
            e = sample_size
        sess.run(train,feed_dict={xp:xy[b:e,0:8],yp:xy[b:e,8:]})
    c_loss = sess.run(loss,feed_dict={xp:xy[:,0:8],yp:xy[:,8:]})
    print(c_loss)
```
